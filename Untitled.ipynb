```python
df_2008 = df_2008.drop([7, 12, 13]).reset_index(drop=True)
```

```python
map_df.rename(columns={'NL_NAME_2': 'City'}, inplace=True)
```

```python
map_df
```

```python
df_2008['City'] = [city.strip() for city in df_2008['City']]
```

```python
df_outer = pd.merge(map_df, df_2008, on='City', how='outer')
```

```python
parties = {'KMT': 'Blues',
               'DPP': 'Greens'}
winning_party = []
for index, row in df_outer.iterrows():
    if row['KMT'] >row['DPP']:
        print(row['KMT'])
        winning_party.append('KMT')
    else:
        winning_party.append('DPP')
df_outer['winning_party'] = winning_party
```

```python
margin = []
for index, row in df_outer.iterrows():
    if row['winning_party'] == 'KMT':
        print(row['KMT'], row['Total'])
        margin.append(row['KMT'] / row['Total'])
    else:
         margin.append(row['DPP'] / row['Total'])
```

```python
df_outer['margin'] = margin
df_2008.at[1, 'winning_party'] = 'DPP'
```

```python
winner = []
for index, row in df_outer.iterrows():
    winner.append(1)
winner[0] = 0.99
winner[16] = 0.99

df_outer['winner'] = winner
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, ax = plt.subplots(figsize=(30, 10))
divider = make_axes_locatable(ax)

for party, color in parties.items():

    cax = divider.append_axes('right', size='2%', pad=0.6)
    cax.set_xlabel(party)

    ax = df_outer[df_outer.winning_party == party].plot(column='margin', cmap=color,
                                         legend=True, ax=ax, cax=cax)
ax.set_title('Taiwan')
```
